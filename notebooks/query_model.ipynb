```python
from openai import OpenAI

client = OpenAI(
    base_url="https://chatqwen3-14-backend.dat-aip-millm.dv.mbcp.cloud/v1",
    api_key="not-needed"  # chave dummy
)
```

```python

stream = client.chat.completions.create(
    model="/app/model/model",
    messages=[{"role": "user", "content": "Escreve um poema curto sobre Lisboa"}],
    stream=True,
)
```

```python
import json, requests

url = "https://chatqwen3-14-backend.dat-aip-millm.dv.mbcp.cloud/v1/chat/completions"
payload = req.json
payload["stream"] = True
with requests.post(url, json=payload, stream=True) as response:
    for line in response.iter_lines():
        #print(line)
        if not line:
            continue
        decoded = line.decode("utf-8").strip()
        if decoded.startswith("data: "):
            decoded = decoded[len("data: "):]
        if decoded == "[DONE]":
            break
        # Cada chunk é um JSON
        chunk = json.loads(decoded)
        delta = chunk["choices"][0]["delta"]
        text = delta.get("content")
        if text:
            print(text, end="", flush=True)
```

```python
import bellmira
```

```python
from bellmira.llm_model.llm_model_client import ModelClient

#client = ModelClient(base_url= "http://host.docker.internal:8888/")
client = ModelClient(base_url= "http://localhost:8888/")
```

```python
client.get_models_name_list()
```

```python
qwen_coder_7b_instruct_req = client.build_chat_request(
    user_prompt="Who are you?",
    system_prompt="",
    model_name='qwen3-30b-instruct-awq-mn',
    temperature=0.7,
    enable_thinking="false"
)
```

```python
reponse = client.send_request(qwen_coder_7b_instruct_req)
reponse.content
```
